```python
# Leitura da da base de dados API
df = yf.download("ADA-USD", start="2019-01-01", end="2021-10-5")
df = df.loc['2021-01-01':'2021-10-5',:]
df.head(2)
```

```python
period = 20
multiplier = 2
```

```python
df['UpperBand'] = df['Close'].rolling(period).mean() + df['Close'].rolling(period).std() * multiplier
df['LowerBand'] = df['Close'].rolling(period).mean() - df['Close'].rolling(period).std() * multiplier
```

```python
plt.rcParams['figure.figsize'] = [18, 8]

plt.rc('font', size=11)

plt.plot(df['Close'], label = "Cardano")
plt.plot(df['UpperBand'], label = "Banda de Bollinger Superior")
plt.plot(df['LowerBand'], label = "Banda de Bollinger Inferior")

plt.legend()

plt.show()
```
